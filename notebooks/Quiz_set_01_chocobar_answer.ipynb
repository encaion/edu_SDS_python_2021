### Q11. df의 index 번호가 짝수인 것만 추출하여 df_sub2 객체에 저장하시오

```python
import numpy as np
```

```python
df_sub2 = df.loc[np.arange(2, len(df), 2), ]
df_sub2.head()
```

```python
df_sub2 = df.loc[2:df.index.max():2, ]
df_sub2.head()
```

### Q12. df_sub1과 df_sub2를 차례대로 병합하여 df_sub 객체에 저장하시오

```python
df_sub = pd.concat([df_sub1, df_sub2])
```
